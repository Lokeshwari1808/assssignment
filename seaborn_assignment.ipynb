```python
"""Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset."""
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
diamonds_data = sns.load_dataset("diamonds")

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple='stack', bins=30, kde=True)


plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Count')


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_matrix = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap using Seaborn
plt.figure(figsize=(15, 10))
sns.heatmap(flights_matrix, annot=True)

# Set plot title and labels
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the heatmap
plt.show()
```
